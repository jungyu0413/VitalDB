```python
# inclusion & exclusion criteria
caseids = list(
    set(df_trks[df_trks['tname'] == 'SNUADC/PLETH']['caseid']) &
    set(df_trks[df_trks['tname'] == 'SNUADC/ECG_II']['caseid']) &
    set(df_cases[df_cases['age'] >= 18]['caseid']) & 
    set(df_cases[df_cases['weight'] >= 30]['caseid']) & 
    set(df_cases[df_cases['weight'] < 140]['caseid']) & 
    set(df_cases[df_cases['height'] >= 135]['caseid']) & 
    set(df_cases[df_cases['height'] < 200]['caseid']) &
    # ex 
    set(df_cases[df_cases['caseend'] > 3600]['caseid']))
print('Total {} cases found'.format(len(caseids)))
np.random.shuffle(caseids)
```

### labeling

* ica

```python
def anocut(x):
  if 1.05 <= x <= 1.35:
    output = 0
  elif x > 1.35:
    output = 1
  else:
    output = -1
  return output
```

```python
df_labs
```

```python
df_labs['label'] = df_labs[df_labs['name'] == 'ica']['result'].map(lambda x : anocut(x))
```

```python
df_labs.head()
```

```python
df_labs_ica = df_labs.dropna()
```

```python
df_labs_ica
```

```python
casid_dn_drop = list(df_labs_ica[(df_labs_ica['dt']) < 0 & (df_labs_ica['label'] == -1.0)]['caseid'])
```

```python
casid_up_drop = list(df_labs_ica[(df_labs_ica['dt']) < 0 & (df_labs_ica['label'] == 1.0)]['caseid'])
```

```python
drop_caseid = casid_dn_drop + casid_up_drop
```

```python
drop_caseid = list(set(drop_caseid))
```

```python
final_ica = final_ica[final_ica['caseid'].map(lambda x : x not in drop_caseid)]
```

* ppg & ecg

```python
dt_list = []
dt_ppg_5 = []
dt_ppg_3 = []
dt_ppg_1 = []
dt_ecg_5 = []
dt_ecg_3 = []
dt_ecg_1 = []
lb_list = []
id = []
for idx in range(len(final_ica)):
    caseid, dt, label = final_ica.iloc[idx]
    vals_ppg_ecg = vitaldb.load_case(int(caseid), ['SNUADC/PLETH','SNUADC/ECG_II'], 1/100)
    vals_dt_ppg = vals_ppg_ecg[:, 0]
    vals_dt_ecg = vals_ppg_ecg[:, 1]
    
    if int(dt) > 0 and int(dt) < len(vals_dt_ppg) and int(dt) > 300 and int(dt) < len(vals_dt_ecg) and int(dt) > 300: 
        dt_list.append(dt)
        dt_ppg_5.append(vals_dt_ppg[int(dt)-300:int(dt)]) 
        dt_ppg_3.append(vals_dt_ppg[int(dt)-180:int(dt)]) 
        dt_ppg_1.append(vals_dt_ppg[int(dt)-60:int(dt)])
        
        dt_ecg_5.append(vals_dt_ecg[int(dt)-300:int(dt)]) 
        dt_ecg_3.append(vals_dt_ecg[int(dt)-180:int(dt)]) 
        dt_ecg_1.append(vals_dt_ecg[int(dt)-60:int(dt)])

        lb_list.append(label)
        id.append(caseid)
        print(caseid)

        if len(id) % 300 == 0:
            np.save('workspace/test_data/ica_id', id)
            np.save('workspace/test_data/ica_lb', lb_list)
            np.save('workspace/test_data/ica_ppg_5', dt_ppg_5)
            np.save('workspace/test_data/ica_ppg_3', dt_ppg_3)
            np.save('workspace/test_data/ica_ppg_1', dt_ppg_1)
            np.save('workspace/test_data/ica_ecg_5', dt_ecg_5)
            np.save('workspace/test_data/ica_ecg_3', dt_ecg_3)
            np.save('workspace/test_data/ica_ecg_1', dt_ecg_1)
            print('save!!')
        else:
            pass

    else:
        pass
```
